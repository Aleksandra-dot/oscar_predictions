```python
import pandas as pd
```

```python
genres_df = pd.read_csv('../data/additional/genres_final.csv')
ratings_df = pd.read_csv('../data/additional/ratings_final.csv').drop('Unnamed: 0', axis=1)
bo_df = pd.read_csv('../data/additional/bo_final.csv').drop('Unnamed: 0', axis=1)
osc_df = pd.read_csv('../data/scraping_results/osc_bp.csv')
```

```python
genres_df.tail(10)
```

```python
merged_df = pd.merge(genres_df, ratings_df, on='wiki', how='inner')

merged_df = merged_df.drop(['year_y', 'film_y'], axis=1).rename(columns={'year_x':'year', 'film_x':'film'})
merged_df.tail(10)
```

```python
merged =  pd.merge(merged_df, bo_df, on=['film', 'year'], how='inner')
merged = merged.drop(['winner_x', 'winner_y', 'wiki_y'], axis=1).rename(columns={'wiki_x':'wiki'}).dropna(subset=['worldwide_box_office'])

#nie ma info o box office All Quiet on the WEstern Front 2022
merged.tail(10)
```

```python
print('Merged df zawiera ', len(merged), ' z ', len(osc_df), ' filmów')
```

```python
merged.tail(10)
```

one hot encoding of genres

```python
genre_dummies = merged['genres'].str.get_dummies(sep=',')
merged_ml = pd.concat([merged[['year', 'film', 'wiki',  'genres', 'rating',
       'numVotes', 'worldwide_box_office']], genre_dummies], axis=1)
merged_ml = merged_ml.drop('genres', axis=1).reset_index().drop(['index', '[]'], axis=1)

#merged_ml zawiera zmienne dotyczące gatunku one-hot oraz rating
merged_ml.to_csv('../data/additional/processed_add.csv')
merged_ml.tail(10)
```

```python
merged_ml.columns
```

```python
#comp_table oscar win + pozostałe konkursy
comp_table = pd.read_csv('../data/processed_results/osc_df')
comp_table = comp_table.drop(['Unnamed: 0', 'winner'], axis=1)
comp_table
```

```python
exclude_columns = ['film', 'year']
full_table= pd.merge(merged_ml,comp_table.drop(exclude_columns, axis=1), on='wiki', how='inner')

#przerabiam box office na zmienną liczbową
full_table['worldwide_box_office'] = full_table['worldwide_box_office'].str.replace('$', '').str.replace(',', '').astype(float).astype(int)

#full table zaweira wszystkie zmienne - rating, genre i konkursy
full_table.to_csv('../data/processed_results/extended_df.csv')
full_table
```
