```python
import pandas as pd
```

```python
genres_df = pd.read_csv('../data/additional/genres_final.csv')
ratings_df = pd.read_csv('../data/additional/ratings_final.csv')
bo_df = pd.read_csv('../data/additional/bo_final.csv')
osc_df = pd.read_csv('../data/scraping_results/osc_bp.csv')
```

```python
# Merge the dataframes on the 'film' column
merged_df = pd.merge(genres_df, ratings_df, on='film', how='inner')
merged_df = pd.merge(merged_df, bo_df, on='film', how='inner')

# Reset the index of the merged dataframe
merged_df.reset_index(drop=True, inplace=True)
merged_df = merged_df.drop_duplicates(subset=['wiki']).drop(['year_x', 'year_y', 'wiki_x', 'wiki_y', 'winner_x', 'winner_y', 'Unnamed: 0_x','Unnamed: 0_y' ], axis=1)
merged_df.to_csv('../data/additional/merged.csv')
merged_df
```

```python
print('Merged df zawiera ', len(merged_df), ' z ', len(osc_df), ' filmów')
```

one hot encoding of genred

```python
genre_dummies = merged_df['genres'].str.get_dummies(sep=',')
merged_ml = pd.concat([merged_df[['year', 'film', 'wiki', 'winner', 'genres', 'rating',
       'numVotes', 'worldwide_box_office']], genre_dummies], axis=1)
merged_ml = merged_ml.drop('genres', axis=1).reset_index()
merged_ml.to_csv('../data/additional/processed_add.csv')
merged_ml.head()
```

```python
noms_table = pd.read_csv('../data/processed_results/osc_df')
noms_table = noms_table.drop('Unnamed: 0', axis=1)
noms_table
```

```python
exclude_columns = ['film', 'winner', 'year']
full_table= pd.merge(merged_ml,noms_table.drop(exclude_columns, axis=1), on='wiki', how='inner').drop(['index', '[]'], axis=1)
full_table.to_csv('../data/processed_results/extended_df.csv')
full_table
```
