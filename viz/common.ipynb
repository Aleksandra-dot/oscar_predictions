```python
# Create an empty dictionary to store the results
results_win = {}
results_nom ={}

# Define the award categories
award_categories = ['BAFTA', 'SAG', 'PGA', 'DGA', 'GG_Drama', 'GG_Comedy','Cannes']  # Add more categories as needed

# Calculate the common nominations and winners for each award category
for category in award_categories:
    # Calculate common nominations
    nom_percentage = (df['nom_' + category.lower()] == 1).mean() * 100
    results_nom[category] = nom_percentage

    # Calculate common winners
    win_percentage = ((df['winner_' + category.lower()] == 1) & (df['Oscar_win'] == 1)).mean() * 100
    results_win[category] = win_percentage

# Convert the results dictionary to a DataFrame
results_win_df = pd.DataFrame.from_dict(results_win, orient='index', columns=['Percentage'])
results_nom_df = pd.DataFrame.from_dict(results_nom, orient='index', columns=['Percentage'])

results_win_df.to_csv('./nominations_win.csv')
results_nom_df.to_csv('./nominations_nom.csv')
# Print the DataFrame
print(results_win_df)
print(results_nom_df)
```
