```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed_results/osc_df')
df = df.drop('Unnamed: 0', axis=1)
df = df[df['year'] > 1995].reset_index().drop('index', axis=1)
df
```

```python
# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
bafta_nom_percentage = (df['nom_bafta'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
bafta_win_percentage = ((df['winner_bafta'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for BAFTA: {:.2f}%".format(bafta_nom_percentage))
print("Percentage of Oscar-winning films also winning BAFTA: {:.2f}%".format(bafta_win_percentage))

# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
gg_drama_nom_percentage = (df['nom_gg_drama'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
gg_drama_win_percentage = ((df['winner_gg_drama'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for Golden Globes Drama: {:.2f}%".format(gg_drama_nom_percentage))
print("Percentage of Oscar-winning films also winning Golden Globes Drama: {:.2f}%".format(gg_drama_win_percentage))

# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
gg_comedy_nom_percentage = (df['nom_gg_comedy'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
gg_comedy_win_percentage = ((df['winner_gg_comedy'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for Golden Globes Comedy: {:.2f}%".format(gg_comedy_nom_percentage))
print("Percentage of Oscar-winning films also winning Golden Globes Comedy: {:.2f}%".format(gg_comedy_win_percentage))

# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
sag_nom_percentage = (df['nom_sag'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
sag_win_percentage = ((df['winner_sag'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for SAG: {:.2f}%".format(sag_nom_percentage))
print("Percentage of Oscar-winning films also winning SAG: {:.2f}%".format(sag_win_percentage))

# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
pga_nom_percentage = (df['nom_pga'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
pga_win_percentage = ((df['winner_pga'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for PGA: {:.2f}%".format(pga_nom_percentage))
print("Percentage of Oscar-winning films also winning PGA: {:.2f}%".format(pga_win_percentage))

# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
dga_nom_percentage = (df['nom_dga'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
dga_win_percentage = ((df['winner_dga'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for DGA: {:.2f}%".format(dga_nom_percentage))
print("Percentage of Oscar-winning films also winning DGA: {:.2f}%".format(dga_win_percentage))

# Obliczenie procentu filmów z kolumny "Film", które mają wartość 1 w kolumnie "bafta_nom"
cannes_nom_percentage = (df['nom_cannes'] == 1).mean() * 100

# Obliczenie procentu filmów z kolumny "Film", które zdobyły Oscara (Oscar_win = 1) i mają wartość 1 w kolumnie "bafta_win"
cannes_win_percentage = ((df['winner_cannes'] == 1) & (df['Oscar_win'] == 1)).mean() * 100

# Wyświetlenie wyników
print("Percentage of films nominated for Cannes: {:.2f}%".format(cannes_nom_percentage))
print("Percentage of Oscar-winning films also winning Cannes: {:.2f}%".format(cannes_win_percentage))
```
